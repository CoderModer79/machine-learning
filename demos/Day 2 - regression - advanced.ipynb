```python
import pandas as pd
import numpy as np
from sklearn import preprocessing, linear_model, model_selection, metrics, pipeline
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/abulbasar/data/master/kaggle-houseprice/data_combined_cleaned.csv")
```

```python
df_clean = df[~df.SalesPrice.isnull()]
```

```python
df_clean.info()
```

```python
del df_clean["Id"]
```

Find the best model based on the rmse score of log value of SalePrice. 
For RMSE score, lower is better.

 - A. Try Lasso model (l1 penalty)
 - B. Try Ridge model (L2 penalty)
 - C. Try Elastic Net model (L1_l2 penalty)

Which algorithm and hyper parameters gives the best result?


```python
target = "SalesPrice"
X = df_clean.copy()
del X[target]
y = np.log(df_clean[target])

X = pd.get_dummies(X, drop_first=True)
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, 
                                        test_size = 0.3, random_state = 1)

est = pipeline.Pipeline([
    ("poly", preprocessing.PolynomialFeatures(degree=1, include_bias=False)),
    ("scaler", preprocessing.StandardScaler()),
    ("lr", linear_model.SGDRegressor(penalty="l1", alpha=0.070, random_state=1, max_iter=1000))
])
est.fit(X_train, y_train)

y_train_pred = est.predict(X_train)
y_test_pred = est.predict(X_test)

print("training r2: %f\ntest r2: %f\ntraining rmse: %f\ntest: rmse: %f" %(
    metrics.r2_score(y_train, y_train_pred),
    metrics.r2_score(y_test, y_test_pred),
    np.sqrt(metrics.mean_squared_error(y_train, y_train_pred)),
    np.sqrt(metrics.mean_squared_error(y_test, y_test_pred))
))

print(est.steps[-1][-1].coef_)
```

```python
target = "SalesPrice"
X = df_clean.copy()
del X[target]
y = np.log(df_clean[target])

X = pd.get_dummies(X, drop_first=True)
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, 
                                        test_size = 0.3, random_state = 1)

est = pipeline.Pipeline([
    ("poly", preprocessing.PolynomialFeatures(degree=1, include_bias=False)),
    ("scaler", preprocessing.StandardScaler()),
    ("lr", linear_model.SGDRegressor(penalty="elasticnet", alpha=0.01
                        , l1_ratio=0.1, tol=1e-6, random_state=1
                        , eta0 = 0.001, max_iter=10000))
])
est.fit(X_train, y_train)

y_train_pred = est.predict(X_train)
y_test_pred = est.predict(X_test)

print("training r2: %f\ntest r2: %f\ntraining rmse: %f\ntest: rmse: %f" %(
    metrics.r2_score(y_train, y_train_pred),
    metrics.r2_score(y_test, y_test_pred),
    np.sqrt(metrics.mean_squared_error(y_train, y_train_pred)),
    np.sqrt(metrics.mean_squared_error(y_test, y_test_pred))
))

model = est.steps[-1][-1] 
print(model.coef_, model.max_iter)
```

```python
scores = model_selection.cross_val_score(est, X_train, 
                            y_train, scoring="neg_mean_squared_error"
                            , cv = 5, verbose=True)
```

```python
cv_score = np.mean(np.sqrt(- scores)) # cv scores based on the rmse
cv_score
```

```python
%%time 

est = pipeline.Pipeline([
    ("poly", preprocessing.PolynomialFeatures(degree=1, include_bias=False)),
    ("scaler", preprocessing.StandardScaler()),
    ("lr", linear_model.SGDRegressor(penalty="elasticnet"
                        , alpha=0.01
                        , l1_ratio=0.1
                        , tol=1e-6
                        , random_state=1
                        , eta0 = 0.001
                        , max_iter=10000))
])
est.fit(X_train, y_train)

param_grid = {
    "lr__l1_ratio": np.random.uniform(0, 0.1, 10),
    "lr__alpha": np.random.uniform(0.3, 0.6, 10)
}
grid = model_selection.GridSearchCV(est, param_grid = param_grid
                                    , verbose=True
                                    , scoring="neg_mean_squared_error"
                                    , cv = 5
                                    , n_jobs = 2
                                   )
grid.fit(X_train, y_train)
print("Best params: ", grid.best_params_)
print("Best rmse score: ", np.sqrt(- grid.best_score_))
```

```python
y_train_pred = grid.predict(X_train)
y_test_pred = grid.predict(X_test)

print("training r2: %f\ntest r2: %f\ntraining rmse: %f\ntest: rmse: %f" %(
    metrics.r2_score(y_train, y_train_pred),
    metrics.r2_score(y_test, y_test_pred),
    np.sqrt(metrics.mean_squared_error(y_train, y_train_pred)),
    np.sqrt(metrics.mean_squared_error(y_test, y_test_pred))
))

model = grid.best_estimator_.steps[-1][-1] 
print(model.coef_, model.max_iter)
```
