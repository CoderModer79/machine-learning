```python
import pandas as pd 
from sklearn import cluster, preprocessing

pd.options.display.max_columns = 1000
```

```python
df = pd.read_csv("/data/snsdata.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
X = df.iloc[:, 4:]
```

```python
X.head()
```

```python
X_std = preprocessing.StandardScaler().fit_transform(X)
kmeans = cluster.KMeans(n_clusters=10, random_state=1)
y = kmeans.fit_predict(X_std)
pd.Series(y).value_counts()
```

```python
df[y == 5].iloc[:, 4:].mean().sort_values(ascending = False) #Sport fans
```

```python
df[y == 5].iloc[:, 4:].std().sort_values(ascending = False) #Sport fans
```
