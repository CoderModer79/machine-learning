```python
mkt_spend_by_state = df.groupby("State")["Marketing Spend"].mean()
mkt_spend_by_state
```

```python
mkt_spend_by_state["New York"]
```

```python
import numpy as np
```

```python
def replace_mkt_spend(row):
    out = row["Marketing Spend"]
    if np.isnan(row["Marketing Spend"]):
        out = mkt_spend_by_state[row.State]
    return out

df["Marketing Spend"] = df.apply(replace_mkt_spend, axis = 1)
df
```

```python
df = df.fillna(df.mean())
df.isnull().sum()
```
