# Top 10 movies

Datasource: https://github.com/abulbasar/data/tree/master/movielens

1. load movies.csv as a dataframe movies
2. load ratings.csv as a dataframe ratings
3. Do you have any null values in movies or ratings
4. Group the ratings datataframe by movieId and find avg rating, and count of ratings
5. Filter out all movieId that have fewer than 100 ratings
6. Sort the data based on the avg rating (descending order)
7. Merge the data with movies to show title
8. Find top 10 movie titles and avg rating

```python
movies = pd.read_csv("/data/movielens/movies.csv")
```

```python
movies.info()
```

```python
ratings = pd.read_csv("/data/movielens/ratings.csv")
```

```python
ratings.info()
```

```python
avg_rating = ratings.groupby("movieId").rating.agg([np.mean, len])
```

```python
avg_rating
```

```python
top10 = avg_rating[avg_rating.len > 100]\
    .sort_values("mean", ascending = False)\
    .iloc[:10, :]
top10
```

```python
top10.merge(movies, left_index=True, right_on="movieId")
```

```python
top10movies = top10.merge(movies, left_index=True, right_on="movieId")[["title", "mean", "movieId"]]
top10movies
```

```python
top10movies.columns = ["Title", "Avg Rating", "Id"]
```

```python
top10movies
```

```python
columns = list(top10movies.columns)
columns[2] = "MovieID"
top10movies.columns = columns
top10movies.head()
```
