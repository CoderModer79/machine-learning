5. Find avg rating for each movie ... group by movieId
6. Exlcude all movies that have got fewer than 100 ratings
7. Find top 10 movies based on the avg rating ... more avg means better
8. Merge the output with movies to show title of the top 10 movies.
9. Final output - movieId, avg_rating, title

```python
top10 = (ratings.groupby("movieId")
 .rating
 .agg(["mean", "count"])
 .query("count > 100")
 .sort_values("mean", ascending = False)
 .iloc[:10, :]
)
top10
```

```python
top10.columns = ["avg_rating", "count"]
top10
```

```python
top10.reset_index().merge(movies
    , on="movieId")[["avg_rating", "title", "movieId"]]
```
