```python
top10 = (ratings
 .groupby("movieId")
 .rating
 .agg(["count", "mean"])
 .query("count >= 100")
 .sort_values("mean", ascending = False)
 .iloc[:10, :]
)
top10 
```

```python
top10.merge(movies, right_on= "movieId", left_index=True)[["title", "movieId", "mean"]]
```
