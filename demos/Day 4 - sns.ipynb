```python
import pandas as pd
from sklearn import *
```

```python
df = pd.read_csv("/data/snsdata.csv")
df.head()
```

```python
df.iloc[:, 4:].sample(10)
```

```python
X = df.iloc[:, 4:].values.astype("float")
features = df.columns[4:]
scaler = preprocessing.StandardScaler()
X = scaler.fit_transform(X)
```

```python
k = 5
```

```python
kmeans = cluster.MiniBatchKMeans(n_clusters=k, random_state=1)
y = kmeans.fit_predict(X)
```

```python
df["cluster"] = y
```

```python
df.cluster.value_counts()
```

```python
centroids = kmeans.cluster_centers_
centroids = pd.DataFrame(scaler.inverse_transform(centroids), columns = features)
centroids.transpose()
```

```python
centroids.iloc[0,:].sort_values(ascending = False)
```

```python
for i in range(k):
    top_terms = centroids.iloc[i,:].sort_values(ascending = False)[:5]
    print("Cluster", i, "\n", top_terms)
    print("\n")
```

```python
features
```

```python
pca = decomposition.PCA()
pca.fit(X)
```

```python
r = pd.Series(pca.explained_variance_ratio_.cumsum())
r[r>0.99]
```

```python
pca = decomposition.PCA(n_components=10)
X_pca = pca.fit_transform(X)
X_pca
```

```python
y_pca = kmeans.fit_predict(X_pca)
```

```python
metrics.homogeneity_score(y, y_pca)
```

```python
df["cluster_pca"] = y_pca
```

```python
df.head()
```

```python
df.cluster_pca.value_counts()
```

```python
for i in range(k):
    print("\nCluster", i, "\n")
    print(df.iloc[:, 4:40][df.cluster_pca == i].mean().sort_values(ascending = False)[:5])
```
